```python
# Get list of subdirs

subdirs_ls = [f.path.split('/')[-1] for f in os.scandir('data') if f.is_dir() ]
subdirs_ls.sort()
subdirs_ls
```

```python
%%time

# NOTE: 

# Face Emtional analysis of upto max_cover_ct *.png images under the data directory

def get_face_emotions(adir):
  """
  Given a subdir of *.png files

  Return a dict[<png filename_root>] = dict[face emo detection factors]
  """
  cover_face_dt = {}
  fail_ct = 0
  success_ct = 0

  max_cover_ct = 1000  # Set upper limit on max number of cover images to process

  # cover_obj_ls = []
  error_log_ls = []
  # cover_date_ls = []

  break_out_fl = False

  total_ct = 0
  for root, subfolders, filenames in os.walk(adir):
    for i,afile in enumerate(filenames):

      if i >= max_cover_ct:
        break_out_fl = True
        break

      if afile.split('.')[-1] == 'png':
        total_ct += 1
        print(f'\nMag Cover #{total_ct} found: {afile} --------------------------------------------------')
        # print(root, subfolders, filenames)
        fullpath2file = f'./{root}/{afile}'
        # print(f'  fullpath: {fullpath2file}')
        try:
          obj = DeepFace.analyze(img_path = fullpath2file, actions = ['age', 'gender', 'race', 'emotion'])
          # cover_obj_ls.append(obj)
          cover_face_dt[afile] = obj
          # cover_date_ls.append(afile.split('.')[0])
          success_ct += 1
          print(f'  <<<SUCCESS>>>')
        except Exception as e:
          error_log_ls.append(f'Failed to upload to ftp: {str(e)}')
          # cover_face_dt[afile] = str(e)
          fail_ct += 1
          print(f'  >>>ERROR<<<: {str(e)}')
            
    if break_out_fl:
      break

  # print(f'DONE: Processed {len(cover_obj_ls)}')
  print(f'DONE: Processed {len(cover_face_dt)}')

  return cover_face_dt, success_ct, fail_ct
```

```python
# Create output subdir if it doesn't already exist

SUBDIR_OUTPUT = 'output'

# Check whether the specified path exists or not
if os.path.exists(SUBDIR_OUTPUT):
  print('subdir exists')
else:
  os.makedirs(SUBDIR_OUTPUT)
  print(f"Created new subdir: {SUBDIR_OUTPUT}\n")

!ls
```

```python
# Write Dictionary to JSON File

from google.colab import files

def save_dict2json(dict_name, json_filename):

  # filename_json_str = f'test_face.json'

  try:
    with open(f'./{SUBDIR_OUTPUT}/{json_filename}', 'w') as fp:
        json.dump(dict_name, fp)
        print(f'SUCCESS: Saved dict:{dict_name} to json_file:{json_filename}\n\n')
        files.download(json_filename)
        return 1
  except:
    print(f"ERROR: Couldn't save dict:{dict_name} to json_file:{json_filename}")
    return -1

# Test

result_code = save_dict2json(test_face_dt, SUBDIR_TEST)

print(f'Saved {test_face_dt}')
print(f'  from subdir: {SUBDIR_TEST}')
print(f'  with result code={result_code}')
```

```python
# Read JSON File into Nested Dictionary

def read_json2dict(json_filename):

  filename_json_str = f'test_face.json'

  try:
    with open(f'./{SUBDIR_OUTPUT}/{json_filename}', 'r') as fp:
      test_read_nested_dt = json.loads(fp.read())
      print(f'SUCCESS: Read json_filename:{json_filename}\n\n')
      return test_read_nested_dt
  except:
    print(f"ERROR: Couldn't read json_file: {json_filename}\n\n")
    return -1

# Test

test_face_fromjson_dt = read_json2dict(SUBDIR_TEST)

if test_face_fromjson_dt == -1:
  print(f'ERROR: Cannot read JSON file: {SUBDIR_TEST}')
else:
  print(f'Read JSON File {SUBDIR_TEST}')
  print(f'  from subdir: {SUBDIR_OUTPUT}\n\n')
  print(json.dumps(test_face_fromjson_dt, sort_keys=True, indent=4))
```
